```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data_diabetes = pd.read_csv('diabetes_prediction_dataset.csv')
```

```python
data_diabetes.head()
```

```python
data_diabetes['diabetes'].value_counts()
```

```python
X = data_diabetes.drop(columns = 'smoking_history', axis = 1)
```

```python
X_data = X.drop(columns = 'gender', axis=1)
```

```python
X_data
```

```python
X_data.groupby('diabetes').mean()
```

```python
x = X_data.iloc[:,:-1].values
```

The colon (:) before the comma means "select all rows."
The :-1 part means "select all columns except the last one." It is using Python's slicing, where -1 refers to the last column, and :-1 excludes it.

```python
x
```

```python
y = X_data.iloc[:,6].values
y
```

.iloc[:, 6]: This is a way to select data using the .iloc indexer:

The colon (:) before the comma indicates that we're selecting all rows.
The number 6 after the comma refers to the column at index 6 (which is the 7th column since indexing starts from 0). This is where the labels or target values (Y) are located.

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y, test_size=0.2, stratify=y, random_state=2)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(X_train, Y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score
print(classification_report(Y_test, y_pred))
print(confusion_matrix(Y_test, y_pred))
```

```python
print("KNN Accuracy:", accuracy_score(Y_test, y_pred))
```
